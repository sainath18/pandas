```python
# concat
d1=pd.DataFrame({"fruit":["apple","banana","orange"],"colour":["red","yellow","orange"]})
d1
```

```python
d2=pd.DataFrame({"fruit":["pineapple","apple","grapes","banana"],"colour":["green","red","black","yellow"],"price":[80,25,50,50]})
d2
```

```python
pd.concat([d1,d2],axis=0)
```

```python
pd.concat([d1,d2],ignore_index=True)
```

```python
# concatting horizontally
pd.concat([d1,d2],axis=1)
```

```python
# merge
m1=pd.DataFrame({"city":["pune","mumbai","banglore","delhi"],"population":[50,180,110,80]})
m1
```

```python
m2=pd.DataFrame({"city":["pune","mumbai","delhi"],"temperature":[42,35,32]})
m2
```

```python
pd.merge(m1,m2,on="city")
```

```python
pd.merge(m1,m2,on="city",how="outer")
```
